```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
```

```python
def compute_module(a):
    return np.sqrt(a.imag**2 + a.real**2)
```

## Cách 1 - dùng scipy

```python
b = [1, 0, -1]
a = [1, 0, 0.95**2]

w1, h1 = signal.freqz(b, a)
```

```python
magnitude1 = compute_module(h1)

fig, ax = plt.subplots(figsize=(20, 4))
ax.plot(w1, magnitude1, color='b')
ax.set_ylabel("Magnitude")
ax.set_xlabel('Normalized Frequency')
ax.grid(True)
ax.set_title('Magnitude response based on Poles and Zeros')
```

## Cách 2 - dùng H(z)

```python
w2 = w1
magnitude2 = np.sqrt(((np.cos(w2) - 1)**2 + np.sin(w2)**2) * ((np.cos(w2) + 1)**2 + np.sin(w2)**2) / (np.cos(w2)**2 + (np.sin(w2) - 0.95)**2) / (np.cos(w2)**2 + (np.sin(w2) + 0.95)**2))
```

```python
fig, ax = plt.subplots(figsize=(20, 4))
ax.plot(w2, magnitude2, color='g')
ax.set_ylabel("Magnitude")
ax.set_xlabel('Normalized Frequency')
ax.grid(True)
ax.set_title('Magnitude response based on H(e^jw)')
plt.show()
```

## Cách 3 - dùng H(e^jw)

```python
w3 = w1
h3 = (1 - np.cos(2 * w3) + 1.j * np.sin(2 * w3)) / (1 + (0.95**2) * (np.cos(2 * w3) - 1.j * np.sin(2 * w3)))
```

```python
magnitude3 = compute_module(h3)
fig, ax = plt.subplots(figsize=(20, 4))

ax.plot(w3, magnitude3, color='r', title='Magnitude response based on H(ejw)')
ax.grid(True)
ax.set_ylabel("Magnitude")
ax.set_xlabel('Normalized Frequency')
ax.set_title('Magnitude response based on freqz(b,a)')
plt.show()
```
